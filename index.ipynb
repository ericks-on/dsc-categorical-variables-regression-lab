## Step 3: Build a Multiple Regression Model with Your Chosen Predictors

Choose the best-looking 3 out of 4 predictors to include in your model.

Make sure that you one-hot encode your categorical predictor(s) (regardless of whether the current data type is a string or number) first.

```python
# Your code here - prepare X and y, including one-hot encoding
y = ames['SalePrice']

X = ames.loc[
    :,
    [continuous_predictor, discrete_number, string_categorical]
].copy()

# one-hot encoding
X = pd.get_dummies(
    X,
    columns=[string_categorical],
    dtype=int,
    drop_first=True
)
X.head()
```

```python
# Your answer here - which category or categories were dropped?
sale_type_categories = set(ames.SaleType.values)

# check dropped
not_dropped = [
    x.split('_')[1] for x in X.columns
    if x.startswith('SaleType')
]

print(
    "Dropped category:",
    [x for x in sale_type_categories if x not in not_dropped]
)
```

```python
# Your code here - build a regression model and display results
import statsmodels.api as sm
model = sm.OLS(y, sm.add_constant(X))

# get results
results = model.fit()

# display results
print(results.summary())
```

## Step 4: Create Partial Regression Plots for Features

For each feature of the regression above (including the dummy features), plot the partial regression.

```python
# Your code here - create partial regression plots
fig = plt.figure(figsize=(20, 15))

sm.graphics.plot_partregress_grid(
    results,
    grid=(3, 4),
    exog_idx=list(X.columns),
    fig=fig
)

plt.tight_layout();
```

## Step 5: Calculate an Error-Based Metric

In addition to the adjusted R-Squared that we can see in the model summary, calculate either MAE or RMSE for this model.

```python
# Your code here - calculate an error-based metric
from sklearn.metrics import mean_absolute_error, mean_squared_error
#get the fitted values 
y_predicted = results.fittedvalues

# get mean absolute error
mae = mean_absolute_error(y, y_predicted)

# get the RMSE
rmse = mean_squared_error(y, y_predicted, squared=False)

print('MAE:', mae)
print('RMSE', rmse)
```

## Level Up (Optional)

Try transforming X using scikit-learn _and_ fitting a scikit-learn linear regression as well. If there are any differences in the result, investigate them.

```python
# Your code here
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(drop='first', sparse_output=False)

X_new = ames.loc[
    :, # all rows
    [continuous_predictor, discrete_number, string_categorical]
].copy().reset_index()

X_new.head()
```

```python
# selecting the categorical column to use one hot encoding
categorical_col = X_new[[string_categorical]].copy()

ohe.fit(categorical_col)

# creating columns for the new transformed
# remember to drop the first
new_columns = [string_categorical + '_' + col for col in ohe.categories_[0][1:]] 

# one hot encoding
ohe_df = pd.DataFrame(
    data=ohe.transform(categorical_col),
    columns=new_columns
)

ohe_df
```

```python
# include the new variables
X_new = pd.concat(
    [X_new.drop(string_categorical, axis=1), ohe_df],
    axis=1
)

X_new.set_index('Id', inplace=True)
X_new.head()
```

```python
# Fitting the model
from sklearn.linear_model import LinearRegression

sk_model = LinearRegression().fit(X_new, y)

# the coefficient of determination
print("sklearn r squared:",sk_model.score(X_new, y))

# printing the coeeficients
print("sklearn coeeficients:")
for i, coef in enumerate(sk_model.coef_):
    print(f"\t{X_new.columns[i]}: {coef}")
```
